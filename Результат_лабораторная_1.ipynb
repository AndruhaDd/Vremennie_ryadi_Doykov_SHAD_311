# Лабораторная работа: анализ и классификация временных рядов

## Цели работы
- Ознакомиться с форматом UCR датасетов (TRAIN/TEST).
- Провести нормировку временных рядов (z-score).
- Визуализировать распределение классов и примеры сигналов.
- Сравнить простейший метод 1-NN с более сложным подходом на признаках (Random Forest).

## Этапы выполнения

### 1. Подготовка данных
Датасеты были загружены в формате UCR: первая колонка — метка класса, остальные — значения временного ряда.  
Для удобства реализована нормализация каждого ряда по формуле:

\[
z_i = \frac{x_i - \mu}{\sigma}
\]

где \(\mu\) — среднее значение ряда, \(\sigma\) — стандартное отклонение.

### 2. Визуализация
- Построены bar-чарты распределения классов для train и test.  
- Для каждого класса показаны несколько случайных рядов, чтобы увидеть вариативность формы.  
- Дополнительно были построены усреднённые ряды с доверительными интервалами (mean ± std).

### 3. Классификация 1-NN
В качестве базового метода использован алгоритм ближайшего соседа (k=1) с евклидовой метрикой.  
Метрика качества — Accuracy и Balanced Accuracy.  
Результаты сохранены и дополнительно визуализированы через confusion matrix.

### 4. Генерация признаков
Для каждого ряда рассчитан набор статистических и структурных характеристик:
- среднее, дисперсия, минимум, максимум, медиана;
- асимметрия и эксцесс;
- энергия сигнала, количество переходов через ноль;
- коэффициент наклона (линейный тренд);
- автокорреляция на лагах 1–5.

### 5. Random Forest по признакам
На построенных признаках обучен Random Forest с подбором параметров по сетке (количество деревьев, глубина).  
Качество оценено на тестовой выборке и сравнивалось с 1-NN.

### 6. Итог
- Проведено сравнение двух подходов:  
  - 1-NN напрямую на рядах (z-нормировка).  
  - Random Forest на признаках.  
- Получена сводная таблица с результатами по каждому датасету.  

## Выводы
- Z-нормировка позволяет корректно сравнивать ряды разных масштабов.  
- Алгоритм 1-NN даёт простую и быструю базу для сравнения.  
- Построение признаков и использование ансамблей (RF) даёт гибкость и зачастую более высокое качество.  
- Визуализация помогает понять характер данных до запуска моделей.
